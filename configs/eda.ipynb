```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("../data/processed/dataset.csv")

print(data.head())
print(data.describe())

# phân bố label
data["label"].value_counts().plot(kind="bar")
plt.title("Label Distribution")
plt.show()

# plot signal mẫu
signal = data.drop("label", axis=1).values[0]

plt.plot(signal)
plt.title("Sample Signal")
plt.show()
```
